```python
data = pd.read_csv("covid_19_country_wise_latest.csv",index_col="Unnamed: 0")
data
```

```python
data.rename(columns={"Unnamed: 0":'Index',"Country/Region":"Country"},inplace=True)
```

```python
def replace_whitespaces(data):
    data.columns = data.columns.str.replace(" ", "_", regex=True)
    return data

data = replace_whitespaces(data)
print("After the white spaces:")
print(data.columns)
```

```python
data.New_cases.mean()
```

```python
data.New_deaths.mean()
```

```python
data.fillna({"New_cases":data.New_cases.mean(),
             "New_deaths":data.New_deaths.mean()},inplace=True)
```

```python
data.isna().sum()
```

```python
#data["Country/Region"].unique()
data.Country.value_counts()
```

```python
data.duplicated(subset=['Country']).sum()
```

```python
data.describe()   
```

```python
data.Deaths.value_counts()
```

```python
data.fillna({"Deaths":data.Deaths[0]},inplace=True)
```

```python
data.Deaths = data.Deaths.str.replace("non","19666").astype(int)
```

```python
data.Deaths = data.Deaths.replace(19666,data.Deaths.mean())
```

```python
data.Deaths.value_counts()
```

```python
data.Deaths.sum(),data.Recovered.sum()
```

```python
data[(data.Confirmed > 1000000)]
```

```python
data[((data.Recovered / data.Confirmed)== 0.95)]
```

```python
data.drop('WHO_Region', axis=1, inplace=True)
data.drop('Confirmed', axis=1, inplace=True)
```

```python
data
```

```python
data.Deaths.max()
data[data.Deaths==data.Deaths.max()]
```

```python
data.sort_values(by='Deaths', ascending=False)
```

```python
data['Total_cases'] = data['Deaths'] + data['Recovered'] + data['Active']
data
```

```python
data['Death_Rate'] = (data['Deaths'] / data['Total_cases']) * 100
data
```

```python
data.Total_cases.mean()
data.Death_Rate.mean()


data[(data.Total_cases > data.Total_cases.mean()) & (data.Death_Rate < data.Death_Rate.mean())]
```

```python
data[['Country', 'Death_Rate']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,15))           #in inches
sns.scatterplot(data=data, x='Total_cases', y='Deaths', hue='Country')
```
